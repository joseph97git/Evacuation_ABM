```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
#data = pd.read_csv('velocities.csv')
data = pd.read_csv('velocity_tangent.csv')
```

```python
data.head()
```

```python
vel = data.iloc[:, 0]
```

```python
x = range(0, len(vel))
```

```python
x_scaled = [val / 100 for val in x]
```

```python
os.getcwd()
```

```python
plt.figure(figsize=(10,10))
sns.set_style("darkgrid")
plt.plot(x_scaled, vel)
plt.title('Horizontal Velocity Component vs. Time', size=25)
plt.ylabel('velocity (m/s)', size=25, labelpad=20)
plt.xlabel('time (s)', size=25, labelpad=20)
plt.xticks(size=20)
plt.yticks(size=20)
#plt.savefig('C:\\Users\\Joseph\\Desktop\\College\\Senior\\Fall\\CAPSTONE\\Agent_based_modeling\\velx_time.eps', format='eps', dpi=1000)
```
